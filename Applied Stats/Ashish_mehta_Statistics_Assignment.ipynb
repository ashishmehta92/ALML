# 2.A :Probability that none of the items are defective?

```python

p = 0.05 # failure rate for the manufacturing items is 5%
n= 10 #number of sample
#This is a binomial distribution case , where we have given probability and Number of samples.

none_of_items_defective = 0;
probability_none_of_items_defective = stats.binom.pmf(none_of_items_defective, n, p)
print(f'Probability that none of the items are defective = {probability_none_of_items_defective}')
```

# 2.B. Probability that exactly one of the items is defective?

```python
exactly_one_item_defective =1
probability_one_item_defective = stats.binom.pmf(exactly_one_item_defective, n, p)
print(f'Probability that exactly one of the items is defective = {probability_one_item_defective}')
```

# 3.C Probability that two or fewer of the items are defective?

```python

#
two_or_fewer_items_defective= 2;
pro_two_or_fewer_items_defective = stats.binom.cdf(two_or_fewer_items_defective, n, p)
print(f'Probability that two or fewer of the items are defective = {pro_two_or_fewer_items_defective}')
```

# 2.D. Probability that three or more of the items are defective ?

```python

# In this we find to find the probability of 2 or fewer items defective and subract it from 1, to calculate the probability three and more
print(f'probability of 3 or fewer items defective  = {1-pro_two_or_fewer_items_defective}')
```

# 3.A. What is Probability that in a given week he will sell some cars?

```python
mean = 3 # mean 3 car per week
# Let say, we will not sell any car and subtract it from 1, to get atleast one car
number_of_car_sell = 0
p_no_car_sell = stats.poisson.pmf(number_of_car_sell, mean) # find the probabilty when he sell no car
p_at_least_sell_some_car = 1- p_no_car_sell;
print(f'Probability that in a given week he will sell some cars = {p_at_least_sell_some_car}')
```

# 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars

```python
# find the commulative density function of 5 and 2 and subtract it.
res = stats.poisson.cdf(4, mean)-stats.poisson.cdf(1, mean)
print(f'Probability that in a given week he will sell 2 or more but less than 5 cars = {res}')
```

# B. Write the null hypothesis and alternative hypothesis.

#### H0 : There is no difference in their mean of funds raised by companies still operating and funds raised by companies that got closed

#### H1: There is a difference in their mean raised by companies still operating and funds raised by companies that got closed

```python
from statsmodels.stats.weightstats import ztest

sample1 =df.loc[df.OperatingState=='Operating','Funds_in_million']
sample2 = df.loc[df.OperatingState=='Closed', 'Funds_in_million']
zscore, p_value = ztest(sample1, sample2)

alpha = 0.05   # Let's consider a significance level of 5%
if p_value < 0.05:
    print(f'Since the p-value {round(p_value,3)} is less than {alpha}  the difference is significant. Hence Rejecting H0 hypothesis')
else:
    print(f'P-value {round(p_value,3)} > alpha {alpha} and  the difference is not significant. Hence Failed to Reject null hypothesis')
```

# There is some difference between the proportion operational companies in the two categories, but let's see if the differnce is significant
There are two ways we can check if the difference is significant
Z-test for proportions
Chi-square test
We shall perform z-test in this case, however, do experiment with Chi-square test and observe if the p-value from both tests gives the same conclusion or not

Null hyputhesis (Ho): The proportion of companies that are operating is the same in both categories - winners and contestants

Alternative hypothesis (Ha): The proportion of companies that are operating is significantly different from each other, among the two categories

# Test for significance and conclusion

```python
from statsmodels.stats.proportion import proportions_ztest

test_statistic, p_value = proportions_ztest([contestants_operating, winners_operating], [contestants, winners])

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```
